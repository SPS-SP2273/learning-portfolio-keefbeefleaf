# A simple plot

## Letâ€™s look at some code.

````python
# Some data for plotting
x = [0, 1, 2, 3, 4, 5, 6]
y_1 = [0, 2, 4, 6, 8, 10, 12]
y_2 = [0, 3, 6, 9, 12, 15, 18]

# Let's start plotting
plt.plot(x, y_1, label='Y values',
         color='red', linestyle='dashed')
plt.xlabel('x-values') 
plt.ylabel('y-values')
plt.title('X vs Y')
plt.grid(alpha=.25)                               # transparency of the grid behind
plt.legend(loc='lower right')                     # location of the plot label, if not present label will not appear
plt.show()
````

### Things to note

````python
plt.plot(x, y_1, color='red', linestyle='dashed', linewidth=2)
````

````python
plt.plot(x, y_1, c='red', ls='dashed', lw=2)
# produces the same plot as above because of the shortforms 
# c - colour
# ls - linestyle
# lw - linewidth
````

GOOD HABIT: include `plt.show()` so as to see an output even in non-interactive environments (i.e. when the code is run directly from the command line)

````python
plt.plot(x, y_1, y_2) # run with default values for all styling parameters
````

Arguments can be split into seperate lines so both is acceptable:
```
plt.plot(x, y, color='red', linestyle='dashed', linewidth=2)
```
```
plt.plot(x, y_1, label='Y values',
           color='red', linestyle='dashed')
```

NOTE: order of the keyword arguments i.e. color does not matter

## Adding another plot

````python
plt.plot(x, y_1, label='Y values',
         color='red', linestyle='dashed')
plt.plot(x, y_2,
         label='Y2 values', color='blue')       # adding another line on to the same graph
plt.xlabel('x-values') 
plt.ylabel('y-values')
plt.title('X vs Y')
plt.grid(alpha=.25)                              
plt.legend(loc='upper left')                     

plt.show()
````

## Yet another plot but with error bars

````python
x = [0, 1, 2, 3, 4, 5, 6]
y_1 = [0, 2, 4, 6, 8, 10, 12]
y_2 = [0, 3, 6, 9, 12, 15, 18]
y_3 = [0, 4, 8, 12, 16, 20, 24]
x_error, y_error = .1, 0.75              # including constant error for all the points 

plt.plot(x, y_1, label='Y1 values', color='red', linestyle='dashed',)
plt.plot(x, y_2, label='Y2 values', color='blue', )
plt.errorbar(x, y_3, xerr=x_error, yerr=y_error,          # adding x and y error bars for the points 
             label='Y3 with errors', color='green')
plt.xlabel('x-values')
plt.ylabel('y-values')
plt.title('X vs Y')
plt.grid(alpha=.25)
plt.legend(loc='upper left')
````

# Better with NumPy

````python
import numpy as np
x = [0, 1, 2, 3, 4, 5, 6]
y_1 = [0, 2, 4, 6, 8, 10, 12]
y_2 = [0, 3, 6, 9, 12, 15, 18]

np_x = np.array(x)
np_y_1 = np.array(y_1)
np_y_2 = np.array(y_2)

plt.plot(np_x, np_y_1, color='red', linestyle='dashed', label='Y values')
plt.plot(np_x, np_y_2, color='blue', label='Y2 values')
````

## Adding mathematical functions

NumPy arrays allow us to generate data-related math functions easily because of the ability to easily transform each element in the NumPy array.

````python
x = np.array([0, 1, 2, 3, 4, 5, 6])

x2 = x**2                  # adding math functions to the arrays
sin_x = np.sin(x)

plt.plot(x, x2, label='x^2',
         color='red', linestyle='dashed', )
plt.plot(x, sin_x, label='sin(x)',
         color='blue')
plt.legend()               # default position is upper left                
````

## We need another axis!

Use `twinx()` command to add another y-axis that shares the same x-axis

````python
x = np.array([0, 1, 2, 3, 4, 5, 6])
x2 = x**2
sin_x = np.sin(x)

plt.plot(x, x2, label='x^2',color='red', linestyle='dashed')
plt.legend(loc='lower left')                                  # for y-axis 1

plt.twinx()                                                   # new y-axis for the plot aft this

plt.plot(x, sin_x, label='sin(x)',color='blue', )
plt.legend(loc='lower right')                                 # for y-axis 2
````

### Things to note

1. there's 2 `legend()` calls now, one for each axis
2. use `np.linspace` to fix the shape of the graph because now there is only a few points now, using:
```
x = np.linspace(starting point, ending point, no. of data points wanted)
```

````python
x = np.linspace(0, 6, 100)                   # getting 100 points instead between x = 0  and  x = 6 
x2 = x**2
sin_x = np.sin(x)


plt.plot(x, x2, label='x^2',color='red', linestyle='dashed')
plt.legend(loc='lower left')                                 

plt.twinx()                                                   

plt.plot(x, sin_x, label='sin(x)',color='blue', )
plt.legend(loc='lower right')    
````

# Saving to disc

use `plt.savefig()` in this format:
```
plt.savefig('destination/path/filename.format', dpi = desired resolution)
```

default file is saved to the folder the notebook is in

````python
x = np.linspace(0, 6, 100)
x2 = x**2
sin_x = np.sin(x)

plt.plot(x, x2, label='x^2',
         color='red', linestyle='dashed')
plt.legend(loc='lower left')
plt.twinx()
plt.plot(x, sin_x,
         label='sin x', color='blue')
plt.legend(loc='lower right')

plt.savefig('simple-plot.png', dpi=150)      # use this to export the plot
````

````python
x = np.linspace(0, 6, 100)
x2 = x**2
sin_x = np.sin(x)

plt.plot(x, x2, label='x^2',
         color='red', linestyle='dashed')
plt.legend(loc='lower left')
plt.twinx()
plt.plot(x, sin_x,
         label='sin x', color='blue')
plt.legend(loc='lower right')

plt.savefig('simple-plot.jpeg', dpi=50)
````

# A real example: Global Warming

## Plotting data from files

````python
data = np.loadtxt('Complete_TAVG_daily.txt',
                   skiprows=24)                        # first few rows don't contain data
date = data[:, 0]                                      
anomaly = data[:, -1]

plt.plot(date, anomaly, alpha=.5)                      # alpha value softens colour of plot
plt.ylim([-3, 3])                                      # sets y limits so as to make y-axis symmetrical
````

````python
data = np.loadtxt('Complete_TAVG_daily.txt',
                   skiprows=24)                      
date = data[:, 0]                                      
anomaly = data[:, -1]

plt.plot(date, anomaly, alpha=.5)                      
plt.ylim([-3, 3])       

plt.hlines(0, date[0], date[-1], linestyle='--', colors='grey')   # adding a horizontal line to show the trend 
                                                                  # y = 0, x1 = first date and x2 = last date
````

## Styles

[Find standard style templates here](https://tonysyu.github.io/raw_content/matplotlib-style-gallery/gallery.html)

````python
plt.style.use('fivethirtyeight')  # using a standard style template

data = np.loadtxt('Complete_TAVG_daily.txt', skiprows=24)

date = data[:, 0]
anomaly = data[:, -1]

plt.plot(date, anomaly, alpha=.5)
plt.hlines(0, date[0], date[-1], linestyle='--', colors='grey')
plt.ylim([-4, 4])

plt.xlabel('Date')
plt.ylabel('Temperature Anomaly')
plt.title('Temperature anomaly\n(Relative to  average from Jan 1951 - Dec 1980.)')
````
