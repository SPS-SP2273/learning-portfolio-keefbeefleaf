# What to expect in this chapter

# 1 Some nomenclature

- figure: the whole canvas where the plots are on
- axis: individual math axes used for plotting, one figure can have multiple axes

# 2 Comparing the two ‘dialects’

```python
import numpy as np
from matplotlib import pyplot as plt

x = np.linspace(-np.pi, np.pi, num=100)
cos_x = np.cos(x)
sin_x = np.sin(x)
```

### pyplot Interface

```python
plt.plot(x, cos_x, label='cos x')
plt.plot(x, sin_x, label='sin x')
plt.legend()
```

### OO Interface

```python
fig, ax = plt.subplots(nrows=1, ncols=1)   # creates a figure and an axis
ax.plot(x, cos_x, label='cos x')
ax.plot(x, sin_x, label='sin x')
ax.legend()
```

# 3 What is OO ax, really?

```python
fig, ax = plt.subplots(nrows=2, ncols=1)    # a figure with 2 rows and 1 column

ax[0].plot(x, cos_x, label='cos x')
ax[1].plot(x, sin_x, label='sin x')

ax[0].legend()
ax[1].legend()
```

```python
type(ax)   # shows `ax` is a NumPy array
```

```python
ax.shape   # `ax` has 2 'things'
```

```python
type(ax[0])   # this is a Matplotlib axis
```

# 4 A complete OO example

```python
fig, ax = plt.subplots(nrows=2, ncols=1,      # 2 axes ready to plot in 2 rows and 1 column
                       figsize=(5, 5),        # change figsize
                       sharex=True)           # whether the plots should share x-axis 
ax[0].plot(x, cos_x, label=r'$\cos(x)$')      # ax is a NumPy array so can be indexed using [0] and [1]
ax[0].fill_between(x, 0, cos_x, alpha=.25)    # fill the plots in
ax[1].plot(x, sin_x, label=r'$\sin(x)$')
ax[1].fill_between(x, 0, sin_x, alpha=.25)

for a in ax:                                  # for loop to iterate through the items in ax
    a.legend()                                # add a legend
    a.grid(alpha=.25)                         # add a grid
    a.set_ylabel('$y$', rotation=0)           # rotate the y-label to horizontal

ax[1].set_xlabel('$x$')                       # only label lowest plot since sharing x-axis which is indexed [1]

fig.suptitle(r'$\sin(x)$ and $\cos(x)$')      # add super title to the FIGURE
fig.tight_layout()                            # make necessary adjustments to make the figure look nice
```

```python
# without fig.tight_layout()
fig, ax = plt.subplots(nrows=2, ncols=1,
                       figsize=(5, 5),
                       sharex=True)
ax[0].plot(x, cos_x, label=r'$\cos(x)$')
ax[0].fill_between(x, 0, cos_x, alpha=.25)
ax[1].plot(x, sin_x, label=r'$\sin(x)$')
ax[1].fill_between(x, 0, sin_x, alpha=.25)

for a in ax:
    a.legend()
    a.grid(alpha=.25)
    a.set_ylabel('$y$', rotation=0)

ax[1].set_xlabel('$x$')

fig.suptitle(r'$\sin(x)$ and $\cos(x)$')
```

# 5 Multiple rows and columns

```python
fig, ax = plt.subplots(nrows=2, ncols=2,
                       figsize=(5, 5),
                       sharex='col', sharey='col')   # the columns share the x and y axes

# Some variables to access the axes and improve readabilty
top_left, top_right, bottom_left, bottom_right = ax.flatten()  # flattens into a 1D array
# [0,0]    [0,1]       [1,0]         [1,1]

top_left.set_title('Top Left')
top_right.set_title('Top Right')
bottom_left.set_title('Bottom Left')
bottom_right.set_title('Bottom Right')

top_left.hlines(y=1, xmin=0, xmax=4)       #`hlines()` function to draw the lines on the plots
top_right.hlines(y=2, xmin=0, xmax=5)
bottom_left.hlines(y=3, xmin=0, xmax=4)
bottom_right.hlines(y=4, xmin=0, xmax=5)

for a in ax.flatten():
    a.grid(alpha=.25)

plt.tight_layout()
```

### Using ax

```python
ax.shape  # shows `ax` is a 2x2 array
```

Can access each of the axes using the following methods:

```python
#1
ax[0, 0].set_title('Top Left')
ax[0, 1].set_title('Top Right')
ax[1, 0].set_title('Bottom Left')
ax[1, 1].set_title('Bottom Right')

#2
top_left=ax[0, 0]
top_right=ax[0, 1]
bottom_left=ax[1, 0]
bottom_right=ax[1, 1]

#3
top_left, top_right, bottom_left, bottom_right = ax.flatten()
```

## 5.1 Sharing axes

```python
fig, ax = plt.subplots(
    nrows=2, ncols=2,
    figsize=(5, 5)       # x and y axes not specified to be shared
)

top_left, top_right, bottom_left, bottom_right = ax.flatten()

top_left.set_title('Top Left')
top_right.set_title('Top Right')
bottom_left.set_title('Bottom Left')
bottom_right.set_title('Bottom Right')

top_left.hlines(y=1, xmin=0, xmax=4)
top_right.hlines(y=2, xmin=0, xmax=5)
bottom_left.hlines(y=3, xmin=0, xmax=4)
bottom_right.hlines(y=4, xmin=0, xmax=5)

for a in ax.flatten():
    a.grid(alpha=.25)
```

```python
fig, ax = plt.subplots(
    nrows=2, ncols=2,
    figsize=(5, 5),
    sharex=True, sharey='row'  # all plots share x axis, y axis is shared across rows
)

top_left, top_right, bottom_left, bottom_right = ax.flatten()

top_left.set_title('Top Left')
top_right.set_title('Top Right')
bottom_left.set_title('Bottom Left')
bottom_right.set_title('Bottom Right')

top_left.hlines(y=1, xmin=0, xmax=4)
top_right.hlines(y=2, xmin=0, xmax=5)
bottom_left.hlines(y=3, xmin=0, xmax=4)
bottom_right.hlines(y=4, xmin=0, xmax=5)

for a in ax.flatten():
    a.grid(alpha=.25)
```

```python
fig, ax = plt.subplots(
    nrows=2, ncols=2,
    figsize=(5, 5),
    sharex='col', sharey='row'  # x axis is shared across columns, y axis is shared across rows
)

top_left, top_right, bottom_left, bottom_right = ax.flatten()

top_left.set_title('Top Left')
top_right.set_title('Top Right')
bottom_left.set_title('Bottom Left')
bottom_right.set_title('Bottom Right')

top_left.hlines(y=1, xmin=0, xmax=4)
top_right.hlines(y=2, xmin=0, xmax=5)
bottom_left.hlines(y=3, xmin=0, xmax=4)
bottom_right.hlines(y=4, xmin=0, xmax=5)

for a in ax.flatten():
    a.grid(alpha=.25)
```

## 5.2 Accessing all axes

To apply changes to all the axes, this can be done:

```python
#1
top_left.grid(alpha=.25)
top_right.grid(alpha=.25)
bottom_left.grid(alpha=.25)
bottom_right.grid(alpha=.25)

#2
for a in ax.flatten():
    a.grid(alpha=.25)
```

# Other useful plots

## Histograms

### Data

### A quick helper function

```python
# creating a function to read the data from the data files easily

def get_plot_data():
    data = {}
    filename = 'sg-gov-graduate-employment-survey_basic_monthly_mean_all.csv'
    data['All'] = np.loadtxt(filename, skiprows=1)        # skip 1st row as it doesn't contain data

    filename = 'sg-gov-graduate-employment-survey_basic_monthly_mean_nus.csv'
    data['NUS'] = np.loadtxt(filename, skiprows=1)

    return data
```

### The histogram

```python
plt.style.use('bmh')    
data = get_plot_data()

plt.hist([data['All'], data['NUS']], bins=50, label=['All', 'NUS']) # bins specifies how many bins to split the data
plt.xlabel('Mean of Basic Montly Earning (S$)')
plt.ylabel('Number of Students')
plt.legend()
```

## Scatter plots

```python
plt.style.use("seaborn-v0_8-darkgrid")

data = get_plot_data()

for label, numbers in data.items():
    x = range(len(numbers))    # to get a list of values for x to plot on the scatter plot
    y = numbers
    plt.scatter(x, y, label=label, alpha=.5)  # to plot a scatter plot

plt.xlabel('Position in the list')
plt.ylabel('Mean of Basic Montly Eraning (S$)')
plt.legend()
```

## Bar charts

```python
student_numbers = {'Life Sciences': 14,
                   'Physics': 12,
                   'Chemistry': 8,
                   'Comp. Biology': 1}       # dummy data
majors = list(student_numbers.keys())        # extracting data from the dictionary
numbers = list(student_numbers.values())

plt.style.use('ggplot')
plt.bar(majors, numbers)                     # create bar graph
plt.xlabel('Majors')
plt.ylabel('Number of Students')
```

```python
student_numbers = {'Life Sciences': 14,
                   'Physics': 12,
                   'Chemistry': 8,
                   'Comp. Biology': 1}      
majors = list(student_numbers.keys())        
numbers = list(student_numbers.values())

plt.style.use('ggplot')
plt.barh(majors, numbers)                     # create HORIZONTAL bar graph
plt.xlabel('Number of students')
plt.ylabel('Majors')
```

## Pie charts

```python
student_numbers = {'Life Sciences': 14,
                   'Physics': 12,
                   'Chemistry': 8,
                   'Comp. Biology': 1}
majors = list(student_numbers.keys())
numbers = list(student_numbers.values())

plt.style.use('fivethirtyeight')
plt.pie(numbers, 
        labels=majors,
        autopct='%1.1f%%',   # format percentages like that, `.1` means 1 decimal place
        startangle=-90                
        )
plt.title('Percentage of each major')
```
