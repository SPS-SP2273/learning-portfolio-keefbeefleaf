# TASK 1

```python
import os

month_to_number = {
'jan': '01', 'feb': '02', 'mar': '03',
'apr': '04', 'may': '05', 'jun': '06',
'jul': '07', 'aug': '08', 'sep': '09',
'oct': '10', 'nov': '11', 'dec': '12'
}

path = '/os-collaboration-exercise-data/oct-08_singapore.txt'
file = path.split(os.path.sep)[-1]
back_name = file.split('-')[-1]     # splits the month name out of the file name
month = file.split('-')[0]
month_number = month_to_number[month] 
corrected_file = f'{month_number}-{back_name}'
print(corrected_file)
```

# TASK 2

```python
def rename_my_file(old_file_name):
    month_to_number = {
'jan': '01', 'feb': '02', 'mar': '03',
'apr': '04', 'may': '05', 'jun': '06',
'jul': '07', 'aug': '08', 'sep': '09',
'oct': '10', 'nov': '11', 'dec': '12'
}
    back_name = old_file_name.split('-')[-1]     
    month = old_file_name.split('-')[0]
    month_number = month_to_number[month] 
    corrected_file = f'{month_number}-{back_name}'
    return corrected_file
```

```python
rename_my_file('may-01_melbourne.txt')
```

# TASK 3 

```python
for old_file_name in ['oct-08_singapore.txt', 'jul-10_cairo.txt', 'may-15_dhaka.txt',
 'may-13_cairo.txt', 'oct-21_cairo.txt', 'jan-10_singapore.txt',
 'jun-20_tokyo.txt', 'aug-06_jakarta.txt', 'dec-21_karachi.txt',
 'jan-01_tokyo.txt']:
    new_file_name = rename_my_file(old_file_name)
    print(f'{old_file_name} ----> {new_file_name}')
```

# TASK 4

```python
glob.glob('os-collaboration-exercise-data/*.txt', recursive=True)
```

```python
import glob

for file_path in glob.glob('os-collaboration-exercise-data/*.txt', recursive=True):
    old_file_name = file_path.split('/')[-1]
    new_file_name = rename_my_file(old_file_name)
    print(f'{old_file_name} ----> {new_file_name}')
```

```python
for old_file_path in glob.glob('os-collaboration-exercise-data/*.txt', recursive=True):
    old_file_name = old_file_path.split('/')[-1]
    new_file_name = rename_my_file(old_file_name)
    new_file_path = os.path.join('os-collaboration-exercise-data', new_file_name)
    os.rename(old_file_path, new_file_path)       # MUST RENAME FILE PATH AS A WHOLE NOT JUST FILE NAME
    print(f'{old_file_name} ----> {new_file_name}')
```

# TASK 5

```python
for country in ['Cairo', 'Dhaka', 'Jakarta', 'Karachi', 'Manila',
 'Melbourne', 'Osaka', 'Shanghai', 'Singapore', 'Tokyo']:
    path = os.path.join('os-collaboration-exercise-data', country)
    os.mkdir(path)
    print(f" Creating {path} folder")
```

```python
import shutil

for cairo_file_path in glob.glob('os-collaboration-exercise-data/*cairo*', recursive=True):
    path_to_folder = 'os-collaboration-exercise-data/Cairo'
    shutil.copy(cairo_file_path, path_to_folder)
    print(f'Copied {cairo_file_path} to {path_to_folder}')
```

# TASK 7

```python
for country_file in glob.glob('os-collaboration-exercise-data/*'):
    country_filetype = os.path.basename(country_file).split('_')[-1]
    country = country_filetype.split('.')[0]
    destination_path = os.path.join('os-collaboration-exercise-data', country)
    shutil.move(country_file, destination_path)
    print(f'"{country_file}" moved to "{destination_path}"')
```
